AGGLOROMERATIVE

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv(r"D:\working\Supervised and Unsupervised learning\unsupervised\clustering\cluster_data.csv")
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(df)
```

```python
k = 2  # Number of clusters
agglo = AgglomerativeClustering(n_clusters=k)
clusters = agglo.fit_predict(data_scaled)
```

```python
plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=clusters, cmap='viridis', marker='o', label='Data Points')
plt.title('Agglomerative Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
